```python
#Defining ticket categories

categories = [
    "Billing",
    "Technical Issue",
    "Account Support",
    "Service Request",
    "General Inquiry"
]
```

```python
# Zero shot classficiation using hugging face model

from transformers import pipeline

classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)

def auto_tag_ticket(text):
    result = classifier(text, categories)
    return result["labels"][0]
```

```python
# Testing with sample tickets

tickets = [
    "I was charged twice this month",
    "Unable to login to my account",
    "Please cancel my subscription",
    "App crashes after update"
]

for t in tickets:
    print(t, "→", auto_tag_ticket(t))
```

```python
# Adding confidence score

def auto_tag_ticket_with_score(text):
    result = classifier(text, categories)
    return {
        "category": result["labels"][0],
        "confidence": round(result["scores"][0], 2)
    }
```

```python
# Deployement using gradio

import gradio as gr

gr.Interface(
    fn=auto_tag_ticket,
    inputs="text",
    outputs="text",
    title="Auto-Tagging Support Tickets",
    description="Automatically assigns categories to support tickets using LLMs"
)

!pip install gradio transformers
!python app.py
```
